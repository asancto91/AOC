```python
import re

def extract_numbers(s):
    """Extract all numbers from a string and concatenate them."""
    return ''.join(re.findall(r'\d+', s))

# Path to your file
file_path = 'AOC1.txt'
# List to store extracted numbers
extracted_numbers = []

# Open and read the file
with open(file_path, 'r') as file:
    for line in file:
        # Extract numbers from each line and add to the list
        numbers = extract_numbers(line)
        extracted_numbers.append(numbers)

# Print or process the extracted numbers
print(extracted_numbers)
```

```python
extracted_chars = []

for number in extracted_numbers:
    # Extract the first and last character
    first_char = number[0]
    last_char = number[-1]

    # Add the pair to the new list
    extracted_chars.append((first_char, last_char))
print(extracted_chars)    
```

```python

# List to store combined numbers
combined_numbers = []

# Iterate over each pair and combine them into numbers
for first_char, last_char in extracted_chars:
    number = int(first_char + last_char)
    combined_numbers.append(number)

# combined_numbers now contains numbers like 25, 99, 33, etc.
print(combined_numbers)
```

```python
# Initialize a variable for the total sum
total_sum = 0

# Iterate over each pair and sum the numbers
for first_char, last_char in extracted_chars:
    total_sum += int(first_char + last_char)

# total_sum now contains the sum of all numbers like 25, 99, 33, etc.
```
