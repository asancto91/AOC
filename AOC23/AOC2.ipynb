```python
import pandas as pd

# Read data from the file
with open('AOC2.txt', 'r') as file:
    data = file.read()

# Split the data into games
games = [game for game in data.split('Game') if ':' in game]

# Prepare a list to hold parsed data
parsed_data = []

# Iterate over each game and parse details
for game in games:
    game_details = game.split(':')
    game_number = game_details[0].strip()  # Game number
    rounds = game_details[1].split(';')  # Split rounds
    rounds = [r.strip() for r in rounds]  # Remove extra spaces
    parsed_data.append([game_number] + rounds)

# Determine the maximum number of rounds
max_rounds = max(len(game) for game in parsed_data)

# Create a DataFrame with dynamic column names
column_names = ["Game Number"] + [f"Round {i}" for i in range(1, max_rounds)]
df = pd.DataFrame(parsed_data, columns=column_names)

print(df)

```

```python

# Function to check if a round meets the conditions
def meets_conditions(round_data):
    if pd.isna(round_data):
        return True  # If the data is NaN (empty round), it's considered as meeting the conditions
    
    # Default color counts
    red_count, green_count, blue_count = 0, 0, 0

    # Parse the color counts from the round data
    color_counts = round_data.split(', ')
    for color_count in color_counts:
        count, color = color_count.split()
        count = int(count)
        if color == 'red':
            red_count = count
        elif color == 'green':
            green_count = count
        elif color == 'blue':
            blue_count = count

    # Check the conditions
    return red_count <= 12 and green_count <= 13 and blue_count <= 14

# Iterate over DataFrame and mark rows that do not meet the conditions
rows_to_delete = set()
for index, row in df.iterrows():
    if not all(meets_conditions(row[col]) for col in df.columns if col.startswith('Round')):
        rows_to_delete.add(index)

# Drop rows that do not meet the conditions
df.drop(rows_to_delete, inplace=True)

export.to_csv(df)

```

```python
import pandas as pd

def extract_color_counts(round_data):
    if pd.isna(round_data):
        return 0, 0, 0  # Return 0 counts for empty rounds
    
    # Parse the color counts from the round data
    counts = {'red': 0, 'green': 0, 'blue': 0}
    color_counts = round_data.split(', ')
    for color_count in color_counts:
        count, color = color_count.split()
        counts[color] = int(count)
    return counts['red'], counts['green'], counts['blue']

# Prepare a DataFrame to hold the maximum color counts for each game
max_colors_df = pd.DataFrame(columns=['Game Number', 'Max Red', 'Max Green', 'Max Blue'])

for index, row in df.iterrows():
    max_red, max_green, max_blue = 0, 0, 0
    for col in df.columns:
        if col.startswith('Round'):
            red_count, green_count, blue_count = extract_color_counts(row[col])
            max_red = max(max_red, red_count)
            max_green = max(max_green, green_count)
            max_blue = max(max_blue, blue_count)
    
    max_colors_df = max_colors_df.append({'Game Number': row['Game Number'], 'Max Red': max_red, 'Max Green': max_green, 'Max Blue': max_blue}, ignore_index=True)

print(max_colors_df)
```
